```python
labels = ['Home','away']
games.HOME_TEAM_WINS.value_counts().plot.pie(autopct="%.1f%%", pctdistance=0.85,labels = labels, cmap='viridis')
plt.title('NBA percentage of win games')
```

```python
box = games[['PTS_home','PTS_away','AST_home','AST_away','REB_home','REB_away']]
```
